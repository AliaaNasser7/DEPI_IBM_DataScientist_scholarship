```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig, ax = plt.subplots()
ax.plot(x, y,'b')
ax.set_xlim(0, 100)
ax.set_ylim(0, 200)
ax.set_yticks(np.arange(0, 201, 50))
plt.show()
```

#### Recreate the plots below using the given data

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])

ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
ax2.set_xticks(np.arange(0, 1.1, 0.2))
ax2.set_yticks(np.arange(0, 1.1, 0.2))
plt.show()
```

Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])
ax1.plot(x,y,'b')
ax1.set_xlim(0, 100)
ax1.set_ylim(0, 200)
ax1.set_yticks(np.arange(0, 201, 50))
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
ax2.plot(x,y,'b')
ax2.set_xticks(np.arange(0, 101, 20))
ax2.set_yticks(np.arange(0, 201, 50))
ax2.set_xlabel('x')
ax2.set_ylabel('y')

plt.show()
```

```python
fig, ax = plt.subplots(1,2)
plt.show()
```

```python
fig, ax = plt.subplots(1,2)
ax[0].plot(x,y,'b--',linewidth=2)
ax[0].set_xlim(0,100)
ax[0].set_ylim(0,200)
ax[0].set_yticks(np.arange(0, 201, 50))

ax[1].plot(x,z,'r',linewidth=2)
ax[1].set_xlim(0,100)
ax[1].set_ylim(0,10000)
plt.show()
```

```python
fig, ax = plt.subplots(1,2,figsize=(16, 3))
ax[0].plot(x,y,'b',linewidth=5)
ax[0].set_xlim(0,100)
ax[0].set_ylim(0,200)
ax[0].set_yticks(np.arange(0, 201, 50))
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')

ax[1].plot(x,z,'r--',linewidth=5)
ax[1].set_xlim(0,100)
ax[1].set_ylim(0,10000)
ax[1].set_xlabel('x')
ax[1].set_ylabel('z')
plt.show()
```

# Seaborn Exercises

## Titanic Data

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.jointplot(x='fare',y='age',data=titanic,kind='scatter', color='#0000ff' ,marginal_kws=dict(color='#4285F4'))
plt.xlim(-100, 600)
plt.ylim(-20, 100)
```

#### Recreate the plots below using the titanic dataframe.

```python
sns.displot(titanic['fare'],bins = 30 , color='lightcoral')
plt.xlim(0, 600)
plt.ylim(0, 500)
plt.show()
```

```python
sns.boxplot(x='class', y='age', data=titanic, palette='rainbow')
plt.ylim(0, 80)
plt.show()
```

```python
sns.countplot(x='sex',data=titanic, palette = {'male': 'blue', 'female': 'green'})
plt.show()
```

```python
matrix = titanic.select_dtypes(include=['int64', 'float64','bool'])
sns.heatmap(matrix.corr(),cmap='coolwarm',annot=False)
plt.title('titanic.corr()')
plt.show()
```
